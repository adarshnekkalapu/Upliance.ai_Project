```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
userDetails= pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx', sheet_name=0)
CookingSessions= pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx', sheet_name=1)
OrderDetails=pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx', sheet_name=2)
```

```python
print(userDetails)
print(CookingSessions)
print(OrderDetails)
```

```python
userDetails['Registration Date'] = pd.to_datetime(userDetails['Registration Date'])
CookingSessions['Session Start'] = pd.to_datetime(CookingSessions['Session Start'])
CookingSessions['Session End'] = pd.to_datetime(CookingSessions['Session End'])
OrderDetails['Order Date'] = pd.to_datetime(OrderDetails['Order Date'])
```

```python
merged_data = pd.merge(userDetails, CookingSessions, on='User ID', how='inner')
merged_data = pd.merge(merged_data, OrderDetails, on='Session ID', how='inner')
```

```python
merged_data
```

```python
merged_data['Rating'] = merged_data['Rating'].fillna(merged_data['Rating'].mean())
```

```python
merged_data
```

```python
df=merged_data
```

```python
df.rename(columns={'User ID_y': 'User ID', 'Meal Type_y': 'Meal Type','Dish Name_y':'Dish Name'}, inplace=True)
```

```python
print(df.describe())
```

```python
popular_dishes_orders = df['Dish Name'].value_counts().head(10)
popular_dishes_cooking = df['Dish Name'].value_counts().head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
popular_dishes_orders.plot(kind='bar', title="Most Popular Dishes (Orders)", figsize=(10, 6))
plt.xlabel('Dish Name')
plt.ylabel('Count')
plt.show()
```

```python
# Which dishes are popular across both orders and cooking sessions?
common_dishes = set(popular_dishes_orders.index).intersection(set(popular_dishes_cooking.index))
print("Common Popular Dishes:", common_dishes)
```

```python
# Checking for a relationship between ratings and order status
rating_status = df.groupby('Order Status')['Rating'].mean()
print("Average Rating by Order Status:", rating_status)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(userDetails['Age'], kde=True, bins=10, color='blue')
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=OrderDetails, x='Meal Type', y='Amount (USD)', palette='Set2')
plt.title('Order Amount Distribution by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Amount (USD)')
plt.show()
```

```python
# Relationship between session duration and session rating
plt.figure(figsize=(8, 6))
sns.scatterplot(data=CookingSessions, x='Duration (mins)', y='Session Rating', color='purple')
plt.title('Session Duration vs Session Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Session Rating')
plt.show()
```

```python
avg_rating_by_location = df.groupby('Location')['Session Rating'].mean().sort_values(ascending=True)
plt.figure(figsize=(10, 6))
avg_rating_by_location.plot(kind='bar', color='teal')
plt.title('Average Session Rating by User Location')
plt.xlabel('Location')
plt.ylabel('Average Session Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot = pd.pivot_table(df, values='Total Orders', index='Age', columns='Favorite Meal', aggfunc='sum', fill_value=0)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, cmap='Blues', fmt='g')
plt.title('Heatmap of User Demographics and Favorite Meals')
plt.xlabel('Favorite Meal')
plt.ylabel('Age')
plt.show()
```
